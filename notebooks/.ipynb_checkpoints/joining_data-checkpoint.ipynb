Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
import numpy as np
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
van_stations_df = pd.read_csv("van_stations.csv")
foursquare_venues_df = pd.read_csv("foursquare_venues.csv")
```

```python
van_stations_df.head()
```

```python
foursquare_venues_df.head()
```

```python
foursquare_venues_df.shape
```

```python
stations_venues_df = pd.merge(van_stations_df, foursquare_venues_df, left_on=["latitude", "longitude"], right_on=["station_latitude", "station_longitude"], how="left")
```

```python
stations_venues_df.head()
```

```python
stations_venues_df.shape
```

```python
stations_venues_df.drop(columns=["latitude", "longitude", "name"], inplace=True)
```

```python
stations_venues_df.head()
```

```python
 # Reorder columns for better readability
column_order = [
        'station_name', 'station_index', 'station_latitude', 'station_longitude', "extra.slots", 
        'venue_name', 'venue_id', 'venue_categories_str',
        'venue_latitude', 'venue_longitude', 'venue_address', 
        'distance_from_station'
    ]
stations_venues_df = stations_venues_df[column_order]
```

```python
stations_venues_df.head()
```

```python
stations_venues_df.shape
```

```python
stations_venues_df.rename(columns={"extra.slots" : "total_slots"}, inplace=True)
```

```python
# Calculate average distance to the station for each station
avg_distance_df = stations_venues_df.groupby('station_index')['distance_from_station'].mean().reset_index()
avg_distance_df = avg_distance_df.merge(
    stations_venues_df[['station_index', 'total_slots']].drop_duplicates(),
    on='station_index'
)
```

```python
# Define category keywords for mapping
restaurants_bars_keywords = ['Restaurant', 'Bar', 'Cafe', 'Coffee', 'Pub', 'Brewery', 'Pizzeria', 'Food', 'Breakfast', 'Bakery', 'Izakaya']
landmarks_museums_keywords = ['Museum', 'Landmark', 'Library', 'Gallery', 'Theatre', 'Monument', 'Park']
banks_shops_keywords = ['Bank', 'Shop', 'Store', 'Market', 'Pharmacy', 'Supermarket', 'Grocery']

def categorize_venue(category_str):
    if pd.isnull(category_str):
        return 'Other'
    if any(kw in category_str for kw in restaurants_bars_keywords):
        return 'Restaurants & Bars'
    if any(kw in category_str for kw in landmarks_museums_keywords):
        return 'Landmarks & Museums'
    if any(kw in category_str for kw in banks_shops_keywords):
        return 'Banks & Shops'
    return 'Other'

stations_venues_df['venue_category_group'] = stations_venues_df['venue_categories_str'].apply(categorize_venue)

venue_counts = stations_venues_df.groupby(['station_index', 'venue_category_group']).size().unstack(fill_value=0)

total_venues_per_station = stations_venues_df.groupby('station_index').size().reset_index(name='total_venues')

station_categories_df = avg_distance_df.merge(venue_counts, on='station_index', how='left')
station_categories_df = station_categories_df.merge(total_venues_per_station, on='station_index', how='left')

column_order = ['station_index', 'distance_from_station', 'total_slots', 'total_venues', 
                'Banks & Shops', 'Landmarks & Museums', 'Other', 'Restaurants & Bars']
station_categories_df = station_categories_df[column_order]

station_categories_df.head()
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
# Visualization: Distribution of venues by distance from station
plt.figure(figsize=(10,6))
stations_venues_df['distance_from_station'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Venue Distances from Station')
plt.xlabel('Distance from Station (meters)')
plt.ylabel('Number of Venues')
plt.grid(False)
plt.savefig("distribution_of_venue_distances.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretation - Distribution of Venue Distances:
The histogram shows a relatively uniform distribution of venues across the 0-300 meter range from bike stations, with the highest concentration occurring at approximately 150-160 meters (340+ venues). This pattern indicates that Vancouver's bike stations are strategically placed in dense commercial corridors rather than isolated locations, as venue counts remain consistently high (280-340 venues) from 100-250 meters. The peak at 150 meters represents an optimal placement zone where stations are close enough for convenience but positioned to serve broader commercial areas. The gradual decline after 250 meters while maintaining substantial venue counts (220+ venues) suggests stations effectively serve entire commercial districts. This uniform distribution demonstrates sophisticated urban planning where bike stations maximize utility by providing consistent access to amenities across their entire 300-meter service radius.

```python
# Additional Visualization: Total venues per station distribution
plt.figure(figsize=(10,6))
station_categories_df['total_venues'].hist(bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Total Venues per Station')
plt.xlabel('Total Venues per Station')
plt.ylabel('Number of Stations')
plt.grid(False)
plt.savefig("total_venues_per_station_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretation - Total Venues per Station Distribution:
The distribution reveals significant variability in venue accessibility across Vancouver's bike stations, with most stations serving between 5-25 nearby venues. A notable outlier exists with one station serving approximately 50 venues, indicating this station is located in an exceptionally dense commercial area. The majority of stations fall into the lower range (5-15 venues), suggesting many stations are positioned in residential or less commercially dense areas. The right-skewed distribution indicates that while most stations provide basic venue access, a few strategic locations serve as major commercial hubs. This pattern reflects Vancouver's mixed urban landscape where some stations serve high-density commercial districts while others provide neighborhood-level access to amenities.

```python
# Box plot showing venue distribution by category
plt.figure(figsize=(12,6))
venue_categories = ['Restaurants & Bars', 'Banks & Shops', 'Other', 'Landmarks & Museums']
venue_data = [station_categories_df[cat] for cat in venue_categories]
plt.boxplot(venue_data, labels=venue_categories)
plt.title('Distribution of Venue Categories Across Stations')
plt.ylabel('Number of Venues')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("venue_categories_boxplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretation - Distribution of Venue Categories:
The box plot reveals that Restaurants & Bars dominate the venue landscape around bike stations, with a median of approximately 11 venues per station and some stations having over 40 restaurant/bar establishments. Banks & Shops and Other categories show similar patterns with medians around 4-6 venues per station and several outlier stations having significantly higher counts. Landmarks & Museums represent the smallest category with most stations having 0-2 cultural venues and a maximum of about 10, indicating limited cultural attraction density. The presence of multiple outliers across all categories suggests that certain bike stations are positioned in exceptionally diverse commercial areas. This distribution confirms that Vancouver's bike-sharing system primarily serves areas rich in dining and social venues, with moderate access to essential services and limited integration with cultural attractions.

```python
# Visualization: Top 10 most common venue categories near stations
top_categories = stations_venues_df['venue_categories_str'].value_counts().head(10)

plt.figure(figsize=(12,8))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 Most Common Venue Categories Near Stations')
plt.xlabel('Number of Venues')
plt.ylabel('Venue Category')
plt.tight_layout()
plt.savefig("top_10_venue_categories.png", dpi=300, bbox_inches='tight')
plt.show()

print("Top 10 Venue Categories with Counts:")
print("=" * 40)
for category, count in top_categories.items():
    print(f"{category}: {count} venues")
    
```

#### Interpretation - Top 10 Venue Categories:
Coffee shops overwhelmingly dominate with 426 venues, representing nearly twice as many locations as the next category, highlighting Vancouver's renowned coffee culture and its integration with bike-sharing for daily routines. The prominence of Asian cuisine (Sushi, Japanese, Vietnamese, Chinese restaurants totaling ~390 venues) showcases Vancouver's multicultural dining scene and significant Asian population influence. All top 10 categories are food-related establishments, indicating that bike stations are strategically positioned to serve Vancouver's vibrant culinary landscape rather than retail or other services. The diversity of specific dining types (pizzeria, café, bakery) suggests bike-sharing users frequently travel for varied food experiences and quick dining options. This pattern demonstrates that Vancouver's bike-share system is deeply integrated with the city's food culture, making it essential for social connectivity and culinary exploration rather than purely utilitarian transportation.

```python
# Additional analysis: Venue density by station slots
plt.figure(figsize=(10,6))
stations_with_slots = station_categories_df[station_categories_df['total_slots'] > 0]
plt.scatter(stations_with_slots['total_slots'], stations_with_slots['total_venues'], 
           alpha=0.6, color='darkblue')
plt.xlabel('Total Bike Slots at Station')
plt.ylabel('Total Venues Near Station')
plt.title('Relationship Between Station Size and Nearby Venues')
plt.grid(True, alpha=0.3)
plt.savefig("station_slots_vs_venues.png", dpi=300, bbox_inches='tight')
plt.show()

# Correlation analysis
correlation = stations_with_slots['total_slots'].corr(stations_with_slots['total_venues'])
print(f"\nCorrelation between station slots and nearby venues: {correlation:.3f}")
```

#### Interpretation - Station Size vs Nearby Venues:
The scatter plot reveals a weak positive correlation (r=0.178) between station size (bike slots) and the number of nearby venues, indicating that larger stations are only moderately associated with higher venue density. Most stations cluster between 15-25 bike slots regardless of venue count, suggesting station sizing decisions are not primarily driven by immediate commercial density. The data shows significant variability, with some small stations (15-20 slots) serving areas with 45-50 venues while some larger stations (30+ slots) serve areas with fewer than 20 venues. This pattern indicates that station capacity planning considers factors beyond just local venue density, such as expected ridership, geographic coverage, or transit connections. The weak correlation suggests that Vancouver's bike-share system prioritizes network coverage and accessibility over simply placing larger stations in the most commercially dense areas.

```python
# Visualization: Total venue counts by category group
plt.figure(figsize=(10,6))
category_totals = station_categories_df[['Banks & Shops', 'Landmarks & Museums', 'Other', 'Restaurants & Bars']].sum()
category_totals.plot(kind='bar')
plt.title('Total Venue Counts by Category Group Across All Stations')
plt.ylabel('Number of Venues')
plt.xlabel('Venue Category Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("total_venue_counts_by_category_group.png", dpi=300, bbox_inches='tight')
plt.show()

# Display the exact numbers
print("Total Venues by Category:")
print("=" * 30)
for category, total in category_totals.items():
    percentage = (total / category_totals.sum()) * 100
    print(f"{category}: {total} venues ({percentage:.1f}%)")

print(f"\nGrand Total: {category_totals.sum()} venues")

```

#### Interpretation - Total Venue Counts by Category Group:
Restaurants & Bars overwhelmingly dominate with approximately 3,700 venues (54.3% of all venues), demonstrating that Vancouver's bike-share system is primarily integrated with the city's dining and social scene. The "Other" category represents the second-largest group with about 1,550 venues (22.3%), likely including professional services, hotels, and miscellaneous businesses that support diverse daily needs. Banks & Shops constitute a moderate presence with 1,327 venues (19.3%), providing essential access to financial services, pharmacies, grocery stores, and retail outlets for routine errands. Landmarks & Museums show minimal representation with only 279 venues (4.1%), suggesting the system serves residents and daily users more than tourists or cultural visitors. This distribution reveals that Vancouver's bike-share system is optimally designed to serve social culture and daily commercial activities, making it an essential component of urban lifestyle rather than primarily a tourist or purely utilitarian service.

```python
# Additional visualization: Average venues per station by category
plt.figure(figsize=(10,6))
category_averages = station_categories_df[['Banks & Shops', 'Landmarks & Museums', 'Other', 'Restaurants & Bars']].mean()
category_averages.plot(kind='bar')
plt.title('Average Number of Venues per Station by Category')
plt.ylabel('Average Venues per Station')
plt.xlabel('Venue Category Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("average_venues_per_station_by_category.png", dpi=300, bbox_inches='tight')
plt.show()

print("\nAverage Venues per Station by Category:")
print("=" * 40)
for category, avg in category_averages.items():
    print(f"{category}: {avg:.2f} venues per station")
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
# Save results to SQLite3 database in data/ directory

def create_connection(path):
    conn = None
    try:
        conn = sqlite3.connect(path)
        print('Connection to SQLite DB successful!')
    except Error as e:
        print(f'The error {e} occurred.')
    return conn
conn = create_connection('../data/city_bikes.sqlite')

van_stations_df.to_sql("van_stations", conn, if_exists="replace", index=False)

foursquare_venues_df.to_sql("foursquare_venues", conn, if_exists="replace", index=False)

stations_venues_df.to_sql("stations_venues", conn, if_exists="replace", index=False)

station_categories_df.to_sql("station_categories", conn, if_exists="replace", index=False)
conn.close()
```
