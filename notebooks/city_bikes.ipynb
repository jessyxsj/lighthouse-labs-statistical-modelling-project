# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd
```

```python
url = 'https://api.citybik.es/v2/networks'
response = requests.get(url)

networks_data = response.json()
df_networks = pd.json_normalize(networks_data['networks'])
df_networks
```

```python
# Search href for Vacouver
df_networks[df_networks['location.city'] == 'Vancouver']
```

```python
# get the stations data in Vancouver
url = 'https://api.citybik.es/v2/networks/mobibikes'
response = requests.get(url)

stations_data = response.json()
df_stations = pd.json_normalize(stations_data['network']['stations'])
df_stations
```

```python
df_stations.shape
```

```python
df_van = df_stations[["name", "latitude", "longitude","extra.slots"]] 
```

```python
df_van.to_csv("van_stations.csv", index=False)
```
